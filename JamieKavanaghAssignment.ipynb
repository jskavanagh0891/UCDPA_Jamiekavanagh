```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm
```

```python
# load the dataset
ElectricVehicles = pd.read_csv('EV Dataset.csv')
```

```python
# Taking a look at the dataset. View the number of rows and columns present in the dataset
ElectricVehicles
```

```python
#Checking for missing Values
missing_values_count = ElectricVehicles.isnull().sum()
missing_values_count
```

```python
#Filling NAs with 0
ElectricVehicles.fillna(0)
```

```python
# To get a better idea of the car manufature I break up the Name column into Brand and Model by splitting by " "
ElectricVehicles[['Brand', 'Model']] = ElectricVehicles['Name'].str.split(' ', 1, expand=True)
```

```python
ElectricVehicles
```

```python
#Capitalzing Names of the Brand to ensure they all start with capitals
ElectricVehicles['Brand'] = ElectricVehicles['Brand'].str.capitalize()
ElectricVehicles
```

```python
#Created a function to class distances by Range
ElectricVehicles.loc[ElectricVehicles['Range']>=500,"Distance"] = 'Long Distance'
ElectricVehicles.loc[ElectricVehicles['Range']<500,"Distance"] = 'Short Distance'
ElectricVehicles.head()
```

```python
# Now I will sort the dataset based on the ranges from highest to lowest
ElectricVehicles.sort_values(by='Range', ascending=False)
```

```python
# Removing Duplicates
Brands = ElectricVehicles['Brand']
Brands.drop_duplicates()
```

```python
# No of unique brands
Brands.nunique()
```

```python
# No. of unique Models
ElectricVehicles['Model'].nunique()
```

```python
# Finding the highest price for a Electric Vehicle
ElectricVehicles['PriceinGermany'].max()
```

```python
# Finding the lowest price while filtering above 1000 in price as some were coming in as 0 and I did not want to include them
EV = ElectricVehicles[(ElectricVehicles['PriceinGermany'] >= 1000)]
EV['PriceinGermany'].min()
```

```python
# Checking the models per brand to see what car manufacturer produces the most amount of electric vehicles
EV = EV.groupby('Brand')['Model'].count()
EV.sort_values(ascending=False)
```

```python
# Checking the top and bottom 5 for Models based on Range
Range= ElectricVehicles.groupby('Model')['Range'].sum()
Range.sort_values(ascending=False)
```

```python
# Checking the brands by price in Germany and sorting by the most expensive
Avg = ElectricVehicles.groupby('Brand')['PriceinGermany'].mean()
Avg.sort_values(ascending=False)
```

```python
# Sorting the average price by Lowest to Highest
Avg.sort_values(ascending=True)
```

```python
# Checking the top 5 and bottom 5 models
Avg_Model = ElectricVehicles.groupby('Model')['PriceinGermany'].sum()
Avg_Model.sort_values(ascending=False)
```

```python
# Graphing the Brands by number of models
EV.sort_values(ascending=True)
EV.plot.barh(figsize=(15,10))
EV.sort_values(ascending=True)
plt.title('Count of models per Car Manufacturer')
plt.xlabel('Brand')
plt.ylabel('No. of Models')
plt.show()
```

```python

Top_Models=Range.nlargest(n=10)
Top_Models_Sorted = Top_Models.sort_values(ascending=True)
Top_Models_Sorted.plot.bar(figsize=(25,10),rot=0,colormap='Accent')
```

```python
Top_Models=Range.nsmallest(n=10)
Top_Models_Sorted = Top_Models.sort_values(ascending=True)
Top_Models_Sorted.plot.bar(figsize=(25,10),rot=0,colormap='Pastel1')
```

```python

Avg_Sorted = Avg.sort_values(ascending=True)
Avg_Sorted.plot.barh(figsize=(15,10))
plt.title('Avg Price per Brand')
plt.xlabel('Price')
plt.ylabel('Brand')
plt.show()
```

```python
Avg_Top = Avg_Model.nlargest(n=10)
Avg_Top_Sorted = Avg_Top.sort_values(ascending=True)
Avg_Top_Sorted.plot.bar(figsize=(25,10),rot=0,colormap='copper')
```

```python
Lowest_Models = ElectricVehicles[(ElectricVehicles['PriceinGermany'] >= 1000)]
Lowest = Lowest_Models.groupby('Model')['PriceinGermany'].sum()
Lowest.sort_values(ascending=False)
```

```python
Avg_Low = Lowest.nsmallest(n=10)
Avg_Low_Sorted = Avg_Low.sort_values(ascending=False)
Avg_Low_Sorted.plot.bar(figsize=(25,10),rot=0,colormap='copper')
```
